```python
def main():
  reglas=[[1,1,5,0.2],[2,2,5,0.7],[3,3,4,0.8],[4,2,4,0.6]]
  porcionGripa=0
  porcionfaringitis=0
  sintomas=[]
  gripa=[]
  faringitis=[]

  tieneSintomas=input("¿Tiene algun sintoma?\n No=0 \n Si=1 \n Opcion ")
  while(tieneSintomas=="1"):
    sintomas.append(int(input("¿Que sintoma tiene?\n 1 fiebre \n 2 Malestar \n 3 Dolor de Garganta \n Sintoma ")))
    tieneSintomas=input("¿Tiene algun otro sintoma?\n No=0 \n Si=1 \n Opcion ")
  

  for sintoma in sintomas:
    for regla in reglas:
      if(sintoma==regla[1] and regla[2]==5):
        gripa.append(regla[3])
      if(sintoma==regla[1] and regla[2]==4):
        faringitis.append(regla[3])

  if len(gripa)==2 or len(gripa)==1:
    if len(gripa)==2:
      porcionGripa=gripa[0]+(1-gripa[0])*gripa[1]
    else:
      porcionGripa=gripa[0]

  if len(faringitis)==2 or len(faringitis)==1:
    if len(faringitis)==2:
      porcionfaringitis=faringitis[0]+(1-faringitis[0])*faringitis[1]
    else:
      porcionfaringitis=faringitis[0]

  if porcionGripa > porcionfaringitis:
    print('Es probable que tenga gripa, la probabilidad es de ',porcionGripa*100,'%')
  if porcionGripa < porcionfaringitis:
    print('Es probable que tenga Faringitis, la probabilidad es de ',porcionfaringitis*100,'%')
  print("Gracias por preferirnos")


main()
```
